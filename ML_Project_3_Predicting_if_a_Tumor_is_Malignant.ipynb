```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
print(breast_cancer_dataset)
```

```python
cancer = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
cancer.head()
```

```python
cancer['label'] = breast_cancer_dataset.target
cancer.tail()
```

```python
print(cancer.shape)
cancer.info()
```

```python
cancer.isnull().sum() #Mentions the missing  values
```

```python
cancer.describe()
```

```python
cancer['label'].value_counts() #1 represents benign, 0 is malignant
```

```python
cancer.groupby('label').mean()
```

```python
X = cancer.drop(columns = 'label', axis = 1)
Y = cancer['label']
print(X)
print(Y)
```

Splitting the data into training and test data

```python
X_train, X_test, Y_train,  Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
print(X.shape, X_train.shape, X_test.shape)
```

Standardizing the Data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train) #
X_test_std = scaler.transform(X_test)
```

Building the Neural Network

```python
import tensorflow as tf
tf.random.set_seed(3) # This helps in maintaining the accuracy score no matter how many times you run the code
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
#setting up the layers of the neural network
model = Sequential([
    keras.layers.Flatten(input_shape = (30,)), #this converts the data into a single dimensional array
    Dense(units = 20, activation = 'relu'),
    Dense(units = 2, activation = 'sigmoid') #No of output units you put is equal to the number of classes you have
])
model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics=['accuracy']
    )
history = model.fit(X_train_std, Y_train, validation_split= 0.1, epochs = 100)
```

Visualizing the Accuracy and Loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Training Data', 'Validation Data'] , loc = 'upper right')
```

Accuracy of the model on Test Data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
Y_predict = model.predict(X_test_std)
print(Y_predict.shape)
print(Y_predict[0])
```

The first value shows the probability of the data being Malignant and second value shows the probability of the data being Benign

```python
print(Y_predict)
```

```python
#converting the prediction probability to class labels (0 or 1)
Y_pred_labels = [np.argmax(i) for i in Y_predict]
print(Y_pred_labels)
```

```python
input_data = (13.08,15.71,85.63,520,0.1075,0.127,0.04568,0.0311,0.1967,0.06811,0.1852,0.7477,1.383,14.67,0.004097,0.01898,0.01698,0.00649,0.01678,0.002425,14.5,20.49,96.09,630.5,0.1312,0.2776,0.189,0.07283,0.3184,0.08183)

def breast_cancer_predict(input_data):
  #change the input data to a numpy array
  input_data_as_numpy_array = np.asarray(input_data)

  #reshape the numpy array as we are predicting one data point
  input_data_reshape = input_data_as_numpy_array.reshape(1,-1)

  #standardizing the input data
  input_data_std = scaler.transform(input_data_reshape)

  prediction = model.predict(input_data_std)
  print(prediction)

  prediction_label = [np.argmax(prediction)]
  print(prediction_label)

  if prediction_label[0] == 0:
    print("The tumor is malignant")
  else:
    print("The tumor is benign")

breast_cancer_predict(input_data)
```
